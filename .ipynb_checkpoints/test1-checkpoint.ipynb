```python
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
```

```python
output_notebook()
```

```python
x = [1, 2, 3, 4, 5]
y = [9, 7, 2, 4, 5]

x1 = [1, 2, 4, 5]
y1 = [9, 7, 4, 5]

#x2 = [1.15, 2.15, 4, 4.9]
#y2 = [9.2, 7.2, 4.35, 5.3]
#x3 = [0.85, 1.85, 4, 5.1]
#y3 = [8.8, 6.7, 3.6, 4.65]

p = figure(plot_width = 800, plot_height = 500, title = "Points")

p.circle(x, y, size = 50, color = "red")
p.line(x, y, line_width = 31, color = "red")
p.circle(x1, y1, size = 30, color = "blue")
p.line(x1, y1, line_width = 31, color = "blue")
#p.line(x2, y2, line_width = 2, color = "blue")
#p.line(x3, y3, line_width = 2, color = "blue")
# show the figure
show(p)
```
